```python
data = np.array([1,2,3,4,5])

mean = np.mean(data)
std = np.std(data)
median = np.median(data)
variance = np.var(data)

normalized_data = (data-mean)/std

print("mean: ", mean)
print("std: ", std)
print("median: ", median)
print("variance: ", variance)

print("normalized_data: ", normalized_data)
```
